## Creation of Coverage Matrix for LSCP & MCLP LINGO Models

```python
#imports
import pandas as pd

#file info
filename_in = 'lscp_model_coverage_matrix_info'
matrix_out = 'lscp_model_coverage_matrix_formatted'
demand_in = 'CensusBlockGroups_demandinfo_acs2018'
```

#### Pandas DF with tabular coverage info. from the ArcGIS Pro spatial join

```python
df = pd.read_csv('{}.csv'.format(filename_in))
print(len(df))
df.head()
```

#### Table with Census Block Group demographic data - HHs with cash/food assistance represent MCLP demand variable "g"

```python
demand = pd.read_csv('C:/Users/jesse/iCloudDrive/SDSU Courses/593 Business Decisions/Brightside/{}.csv'.format(demand_in))
print(len(demand))
demand.head()
```

```python
# merge main df with demand df
df = df.merge(demand, how = 'left', on = 'GEOID')
print(len(df))
df.head()
```

#### Create lists for LSCP, MCLP variables and data - c (candidates) for sites, bg (block groups) for demand units, g (hhs w/ snap) for demand

```python
#list of candidate ids (join_fid) - aij matrix columns
c = df.JOIN_FID.unique()
c.sort()
print(len(c))
print(c)
```

```python
#list of block group geoids - aij matrix rows
bg = df.GEOID.unique()
bg.sort()
print(len(bg))
print(bg)
```

```python
# list of HHs with SNAP (for MCLP demand)
g = []

for group in bg:
    temp = df.query('GEOID == {}'.format(group))
    num = temp['with_cash_snap_assistance'].unique()
    g.append(num[0])
    
print(len(g))
print(g)
```

#### Create Aij Coverage Matrix

```python
#create empty aij matrix (all values zero)
aij = pd.DataFrame()

#add row for each block group
aij['GEOID'] = bg

#add column for each candidate
for candidate in c:
    aij['{}'.format(candidate)] = 0

print(len(aij))
aij.head()
```

```python
#fill in aij matrix
for i, row in df.iterrows():
    #store geoid, candidate values
    geoid = df['GEOID'][i]
    candidate = df['JOIN_FID'][i]
    
    #get index of GEOID row in aij
    idx = aij[aij['GEOID']==geoid].index.values[0]
    
    #value = 1 if geoid is covered by candidate
    aij.loc[idx, '{}'.format(candidate)] = 1
    
    
print(len(aij))
aij.head()
```

```python
#double check aij matrix - sum of all 1s should equal number of rows in original table (df)
matches = 0

for candidate in c:
    
    temp = aij['{}'.format(candidate)].sum()
    matches = matches + temp
    
print(matches)
```

```python
#save coverage matrix as csv
aij.to_csv('{}.csv'.format(matrix_out), index=False)
```
